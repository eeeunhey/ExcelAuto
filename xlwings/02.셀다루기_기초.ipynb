```python
# Q. 교육이수현황.xlsx을 열고, 마케팅팀 시트 선택
```

```python

import xlwings as xw
# xw.app (소문자) -> xw.App (대문자)
# app_books -> add_book
app = xw.App(add_book=False)
```

```python
wb = app.books.open('교육이수현황.xlsx')
```

```python
ws = wb.sheets['마케팅팀']
```

```python
# 해당 셀 데이터 넣기
ws.range('B2').value = "성명"
```

```python
ws.range('C2').value = "1월"
```

```python
ws.range('D2').value = "코드"
```

```python
ws.range('B3').value = "은혜"
ws.range('C3').value = 12
```

```python
# 데이터 여러개 입력하기

ws.range('F2').value = [['부서','코드'],['마케팅팀', 'MKT01'],['영업1팀','SEL01'],['영업2팀','SEL02']]
```

```python
ws.range('A3').options(transpose = True).value = ['마케팅팀','영업1팀', '영업2팀']
```

```python
# B4셀부터 아래로 값이 채워짐
ws.range('B4').options(transpose = True).value = ['은지','민희']
```

```python
ws.range('C4').options(transpose = True).value = ['12','12']
```

```python
# 합계 구하기 
# SUM 수식 입력하기

ws.range('B6').value = '합계'
ws.range('C6').value = '=SUM(C3:C5)'
```

```python
ws.range('D3').options(transpose = True).value = '=VLOOKUP(A3,$F$2:$G$5,2)'
```

```python
# 셀 한개 접근하기
ws.range('A1')
```

```python
# 셀 한개 값 가져오가
 
ws.range('B2').value
```
