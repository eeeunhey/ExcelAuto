```python
import xlwings as xw

# 엑셀 앱 만들기
# 엑셀 실행 시 빈 파일을 실행하기 위해 Fasle 로 지정한다
# Fasle 실행하면 시트가 없느 빈 화면이 실행한다
# True 나 () 이런식으로 작성하면 엑셀 시트가 생성된채로 실행된다
app = xw.App(add_book=False)

app
```

```python
# 엑셀 워크북 만들기
# 시트 생성이 된다
wb = app.books.add()
wb
```

```python
# 워크 시트 할당하기 
# 이렇게 변수명에 저장해서 이 ws를 호출하면 이름 변경 뭐 여러 작업이 가능하다
ws =  wb.sheets['Sheet1']
ws
```

```python
ws.name = "영업1팀"
```

```python
ws = wb.sheets['Sheet2']
ws.name = "영업2팀"
```

```python
# 다른 이름으로 엑셀 저장
wb.save('교육이수현황.xlsx')
```

```python
# 엑셀 시트가 모두 닫힌다
wb.close()
```

```python
# 엑셀 앱 닫기
app.quit()
```

```python
# 엑셀 실행 시킨다음에 가능하다
app = xw.App(add_book=False)
```

```python
# 엑셀 워크북 불러오기
wb = app.books.open('교육이수현황.xlsx')
```

```python
# 새로운 시트 생성하기
wb.sheets.add('영업3팀')
wb.sheets.add('마케팅팀')
```

```python
# 시트를 추가하는 이름 지정하고 before / after =(전)해당시트(후)
# 위치를 지정해서 시트를 생성할 수 있다   
wb.sheets.add(name='세무회계팀', before='영업2팀')
```

```python
# 워크 시트 선택(인덱스)
# 0부터 시작한
wb.sheets[2]  
```

```python
# 워크 시트 내용 지우기
wb.sheets[2].clear()
```

```python
# 워크 시트 삭제
wb.sheets[2].delete()
```

```python
# 시트 활성화
# A1번에 셀 선택된다
wb.sheets['마케팅팀'].activate()
```

```python
# 엑셀 저장
wb.save()
```
